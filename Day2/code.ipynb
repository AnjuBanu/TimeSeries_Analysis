```python
from dateutil.parser import parse 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
fig, axes = plt.subplots(1,3,figsize=(15,4), dpi=100)

data1 =  pd.read_csv("../Data/sunspotarea.csv",parse_dates=["date"], index_col='date')
data2 =  pd.read_csv("../Data/guinearice.csv",parse_dates=["date"], index_col='date')
data3 =  pd.read_csv("../Data/AirPassenger.csv",parse_dates=["Months"],index_col='Months')

data1.plot(ax=axes[0],title="Seasonality Only")
data2.plot(ax=axes[1],title="Trend Only")
data3.plot(ax=axes[2],title="Seasonality and Trend")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
df =  pd.read_csv("../Data/a10.csv",parse_dates=["date"], index_col='date')
add_decompose = seasonal_decompose(df['value'], model='additive',extrapolate_trend='freq')
mul_decompose = seasonal_decompose(df['value'], model='multiplicative',extrapolate_trend='freq')

add_decompose.plot().suptitle('Additive Decompose', fontsize=22)
mul_decompose.plot().suptitle('Multiplicative Decompose', fontsize=22)
plt.tight_layout()
```
